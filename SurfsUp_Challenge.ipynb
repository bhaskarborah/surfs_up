```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

import numpy as np
import pandas as pd
```

```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temperatures = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date) == '06').all()
```

```python
#  3. Convert the June temperatures to a list.
june_temperatures_list = list(june_temperatures)
june_temperatures_list
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temperatures_df = pd.DataFrame(june_temperatures_list, columns=['date','June Temps'])
june_temperatures_df.set_index(june_temperatures_df['date'], inplace=True)
june_temperatures_df=june_temperatures_df[['June Temps']]
june_temperatures_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temperatures_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_temperatures = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date) == '12').all()
```

```python
# 7. Convert the December temperatures to a list.
december_temperatures_list = list(december_temperatures)
december_temperatures_list
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_temperatures_df = pd.DataFrame(december_temperatures_list, columns=['date','December Temps'])
december_temperatures_df.set_index(december_temperatures_df['date'], inplace=True)
december_temperatures_df=december_temperatures_df[['December Temps']]
december_temperatures_df
```

```python
# 9. Calculate and print out the summary statistics for the December temperature DataFrame.
december_temperatures_df.describe()
```

```python
prec_june = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == '06').all()
df_june = pd.DataFrame(prec_june, columns=['date','precipitation'])
df_june.set_index(df_june['date'], inplace=True)
df_june = df_june.sort_index()
df_june.plot(rot=90)
plt.xticks()
plt.show()
```
